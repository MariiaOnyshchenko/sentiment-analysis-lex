```python
import pymorphy2
import re
morph = pymorphy2.MorphAnalyzer(lang='uk')
```

```python
user_text = input()
clean_text = re.sub(r'[^\w\s]|[\d]', '', user_text)
word_list = clean_text.split()
print (len(word_list), word_list)
print (len(set(word_list)))
normal_word_list=[]
for i in word_list:
    normal_word_list.append(morph.parse(i)[0].normal_form)
    print(morph.parse(i)[0])
print (len(set(normal_word_list)))
print (normal_word_list)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
ex_data = pd.read_excel('DictCopy.xlsx')


print (ex_data[:10])
```

```python
df_ordered = pd.concat([ex_data[ex_data['Слово//словосполучення']==x] for x in normal_word_list]) #динаміка тексту
df_ordered
```

```python
print(normal_word_list)
recognized_words = df_ordered['Слово//словосполучення'].values.tolist()
print(recognized_words)
for i in range (1, len(normal_word_list)-1):
    if normal_word_list[i-1]=='не':
        if normal_word_list[i] in recognized_words:
            number = df_ordered[df_ordered['Слово//словосполучення']==normal_word_list[i]].index.values
            print (number)
            target = df_ordered.loc[number, 'Середня оцінка тональності']
            print (target.iloc[0])
            df_ordered.loc[number[0], 'Середня оцінка тональності'] = -target.iloc[0]
df_ordered
```

```python
df_ordered.plot.bar(x = 'Слово//словосполучення', y='Середня оцінка тональності', facecolor="lightgreen", figsize=(20,10))

plt.xticks(ticks = [i for i in range (len(df_ordered))], labels=df_ordered['Слово//словосполучення'], rotation="45", fontsize=12)
plt.yticks(fontsize=12)

mean_eval = round(df_ordered['Середня оцінка тональності'].mean(), 3)
print (f"Загальна оцінка для всього тексту: {mean_eval}")
```
